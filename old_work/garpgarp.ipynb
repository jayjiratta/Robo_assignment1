```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
combined_0_75 = pd.read_csv('combined_2_(speed 0.75).csv')
print(combined_0_75)
```

```python
combined_1_5 = pd.read_csv('combined_2_(speed 1.5).csv')
print(combined_1_5)
```

```python
plt.plot(combined_1_5['times(seconds)'], combined_1_5['tof1'], label='speed 1.5 : TOF1')
plt.plot(combined_0_75['times(seconds)'], combined_0_75['tof1'], label='speed 0.75 : TOF1')

plt.xlabel('Time (seconds)')
plt.ylabel('TOF')
plt.title('TOF1')
plt.legend()

plt.show()
```

```python
plt.plot(combined_1_5['times(seconds)'], combined_1_5['gyro_z'], label='speed 1.5 : gyro_z')
plt.plot(combined_0_75['times(seconds)'], combined_0_75['gyro_z'], label='speed 0.75 : gyro_z')

plt.xlabel('Time (seconds)')
plt.ylabel('gyro_z')
plt.title('gyro_z')
plt.legend()

plt.show()
```
